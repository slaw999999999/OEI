```python
P1 = [-7430,-6310,5180,5080,4080,5420,5770,5770,4940,5020]
P2 = [-6170,-6680,4220,5450,5030,4820,4870,4020,5400,5320]
P3 = [-6520,-5830,4680,5380,4490,4970,5060,4360,4110,5890]
P4 = [-6280,-6780,4060,4130,4880,4500,4300,5030,5870,5400]

STOPA_DYSKONTOWA = 0.15
```

NPV

```python
WYNIK = 0
PLOT_YEARS = []
PLOT_VALUES = []
for y,x in enumerate(P1):
    WYNIK = WYNIK + (x)/((1+STOPA_DYSKONTOWA)**(y))
    PLOT_YEARS.append(y)
    PLOT_VALUES.append(WYNIK)
print(f"Twoje NPV dla inwestycji 1 wynosi: {WYNIK:.2f}")

# WIZUALIZACJA GRAFICZNA
import pandas as pd
import matplotlib.pyplot as plt

d = {'x' : PLOT_YEARS,
     'zł' : PLOT_VALUES}

df = pd.DataFrame(d)

df.plot('x',y=['zł'])
plt.show()
```
